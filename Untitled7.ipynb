```python
from tmdbv3api import TMDb, Movie, TV, Discover
from datetime import datetime
import random
import time
import sys

# --- CONFIGURATION ---
tmdb = TMDb()
tmdb.api_key = 'a3ce43541791ff5e752a8e62ce0fcde2'
tmdb.language = 'en'

movie_api = Movie()
tv_api = TV()
discover_api = Discover()

# --- SAFETY FILTERS ---
BANNED_KEYWORDS = ['erotic', 'sexual', 'gore', 'slasher', 'adult', 'crime scene']

def is_safe(item, is_adult):
    """Strictly filters out adult content and mature genres for minors."""
    if is_adult: return True
    if getattr(item, 'adult', False): return False

    # Block Horror (27) and Thriller (53) for under 18
    genre_ids = getattr(item, 'genre_ids', [])
    if any(gid in [27, 53] for gid in genre_ids): return False

    text = (getattr(item, 'title', '') + getattr(item, 'overview', '')).lower()
    return not any(word in text for word in BANNED_KEYWORDS)

def get_ott(media_id, m_type):
    """Fetches the OTT provider for India."""
    try:
        details = movie_api.details(media_id, append_to_response="watch/providers") if m_type == 'movie' else tv_api.details(media_id, append_to_response="watch/providers")
        return details.get('watch/providers', {}).get('results', {}).get('IN', {}).get('flatrate', [{}])[0].get('provider_name', "Theaters/Rent")
    except: return "Unknown"

def process_session():
    print("\n" + "‚ïê"*50 + "\nüé¨ SECURE MOVIE FINDER (Released Only)\n" + "‚ïê"*50)

    name = input("Name: ").strip().title()
    try:
        age = int(input("Age: "))
    except:
        age = 18
    is_adult = age >= 18

    m_type = 'tv' if input("\n1. Movies | 2. Web Series: ").strip() == '2' else 'movie'

    print("\nüó£Ô∏è Language: 1.Telugu 2.Hindi 3.Tamil 4.English 5.Malayalam 6.Kannada")
    langs = {"1":"te", "2":"hi", "3":"ta", "4":"en", "5":"ml", "6":"kn"}
    sel_lang = langs.get(input("Select Number: "), "en")

    mood_map = {"happy": 35, "sad": 18, "excited": 28, "scared": 27, "bore": 53}
    mood = input("\nüé≠ Mood (happy/sad/excited/scared/bore): ").lower()
    g_id = mood_map.get(mood, 35)

    print("\nüîç Searching for released content...")
    today = datetime.now().strftime('%Y-%m-%d')

    results_pool = []
    # Fetching up to 5 pages (100 movies) to ensure we find 50 safe, released ones
    for p in range(1, 6):
        params = {
            'with_genres': g_id,
            'with_original_language': sel_lang,
            'page': p,
            'include_adult': is_adult,
            'primary_release_date.lte': today, # üü¢ ONLY RELEASED MOVIES
            'first_air_date.lte': today,       # üü¢ ONLY RELEASED SERIES
            'watch_region': 'IN'
        }
        data = discover_api.discover_movies(params) if m_type == 'movie' else discover_api.discover_tv_shows(params)
        results_pool.extend(data)

    # Filter by Safety and ensure they have a description
    final_list = [m for m in results_pool if is_safe(m, is_adult) and m.overview][:50]

    if not final_list:
        print("‚ùå No released content found for this mood/language.")
        return

    # --- NAVIGATION LOOP ---
    idx = 0
    while idx < len(final_list):
        current_page = (idx // 10) + 1
        print(f"\n--- Page {current_page} of 5 ---")
        print(f"{'ID':<4} {'Title':<35} {'Rating':<6} {'OTT'}")
        print("-" * 65)

        chunk = final_list[idx : idx + 10]
        for i, m in enumerate(chunk):
            title = getattr(m, 'title', getattr(m, 'name', 'Unknown'))
            rating = str(m.vote_average)[:3]
            print(f"{idx+i+1:<4} {title[:33]:<35} {rating:<6} {get_ott(m.id, m_type)}")

        print("-" * 65)
        print("Commands: [Number] Info | 'n' Next | 'b' Back | '0' Exit")
        cmd = input("üëâ Enter choice: ").strip().lower()

        if cmd == 'n':
            if idx + 10 < len(final_list): idx += 10
            else: print("\nüèÅ No more results.")
        elif cmd == 'b':
            if idx - 10 >= 0: idx -= 10
            else: print("\nüè† You are on page 1.")
        elif cmd == '0':
            break
        elif cmd.isdigit():
            choice = int(cmd)
            if 1 <= choice <= len(final_list):
                item = final_list[choice-1]
                date = getattr(item, 'release_date', getattr(item, 'first_air_date', 'N/A'))
                print(f"\nüì¢ {getattr(item, 'title', getattr(item, 'name', ''))} ({date[:4]})")
                print(f"üìù PLOT: {item.overview}")
                input("\n[Press Enter to return to list]")
            else:
                print("‚ùå Invalid number.")
        else:
            print("‚ùå Invalid command.")

if __name__ == "__main__":
    while True:
        process_session()
```
